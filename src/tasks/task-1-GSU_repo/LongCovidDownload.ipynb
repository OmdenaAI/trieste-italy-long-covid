```python
import gzip
import shutil
import os
import wget
import csv
import linecache
from shutil import copyfile
import ipywidgets as widgets
import numpy as np
import pandas as pd
import snscrape
import snscrape.modules.twitter as sntwitter
```

```python
def hydrate_tweet(tweet_id, tweets):
    #print(tweet_id)
    tweets_list = []
    # Using TwitterSearchScraper to scrape data and append tweets to list
    try:
        for i, tweet in enumerate(snscrape.modules.twitter.TwitterTweetScraper(tweetId=tweet_id,
                                                                mode=snscrape.modules.twitter.TwitterTweetScraperMode.SINGLE).get_items()):
            tweets.append([tweet.date, tweet.user.username, tweet_id, tweet.content])
    except:
        pass
```

```python
dataset_URL = "https://github.com/thepanacealab/covid19_twitter/blob/master/dailies/2021-01-20/2021-01-20_clean-dataset.tsv.gz?raw=true" #@param {type:"string"}

out = os.path.basename(dataset_URL).replace('?raw=true', '')
```

```python

urls = [dataset_URL]
for ds in urls:
    tweets = []
    out = os.path.basename(ds).replace('?raw=true', '')
    wget.download(ds, out = out)
    #Unzips the dataset and gets the TSV dataset
    with gzip.open(out, 'rb') as f_in:
        with open(out[:-3], 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
            #Deletes the compressed GZ file
            #os.unlink(out)
              #Gets all possible languages from the dataset
            df = pd.read_csv(out[:-3],sep="\t")
            df = df.dropna()
            #filtering for region and language
            df = df[(df['lang']=='it') & (df['country_code']=='IT')]
            print(df.shape)
            ids = df['tweet_id'].values
            for id in ids:
                print(id)
                hydrate_tweet(id, tweets)
                #break
            hdf = pd.DataFrame(tweets, columns=['Date', 'User','Tweet_id', 'Tweet'])
            hdf.to_csv('C:/LongCovidData/hydrated_'+out[:-7]+'.csv', index=False)
```
