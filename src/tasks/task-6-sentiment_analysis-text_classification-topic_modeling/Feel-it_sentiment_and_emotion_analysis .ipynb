```python
dfs = [data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12]
```

```python
for i, df in tqdm(enumerate(dfs)):
    df['emotion'] = emotion_classifier.predict(df.tweet.to_list())
    print(df.head)
    df.to_csv(f'data_{i}.csv')
```
