```python
ser = []

ser.append(pd.read_json('/content/tokens_freq_2021_Jul_Oct.json', typ='series'))
ser.append(pd.read_json('/content/tokens_freq_Nov2021_Jan2022.json', typ='series'))
ser.append(pd.read_json('/content/tokens_freq_2022_Feb_Apr.json', typ='series'))
ser.append(pd.read_json('/content/tokens_freq_2022_May_Jul.json', typ='series'))
```

```python
keyword = 'stanchezza'

x = np.arange(2,10,2)

i=0
y = []
while i < len(ser):
  try:
    y.append(ser[i][keyword])
    print(f'{i+1} quarter: ', ser[i][keyword])
    i += 1
  except:
    y.append(0.0)
    i += 1
    print(f'{i+1} quarter: ', 'no value')
  

y = np.array(y)
#print(y)

#y = np.array([ser1[keyword], ser2[keyword], ser3[keyword], ser4[keyword]])


x_ticks_labels = ['1st quarter','2nd quarter','3rd quarter','4th quarter']

fig, ax = plt.subplots(1,1) 
ax.plot(x,y)

plt.title(f"keyword '{keyword}' quartely analysis")
# Set number of ticks for x-axis
ax.set_xticks(x)
# Set ticks labels for x-axis
ax.set_xticklabels(x_ticks_labels, rotation='vertical', fontsize=12)
plt.show()
```
